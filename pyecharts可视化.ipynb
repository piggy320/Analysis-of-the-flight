各航空公司占比——饼图

```python
import pandas as pd
import numpy as np
from pyecharts.charts import Line#折线图所导入的包
from pyecharts.charts import Bar#柱状图所导入的包
from pyecharts.charts import Pie#饼图所导入的包
from pyecharts import options as opts#全局设置所导入的包
from pyecharts.charts import Bar, Grid, Line,WordCloud
from pyecharts.options.global_options import ThemeType
from pyecharts.globals import SymbolType

def setpie(): 
    #统计各航空公司所以航班数量
    data = pd.read_csv(r"C:\Users\54127\Desktop\同程旅游k.csv",encoding='utf-8')
    d1=data.loc[:,'flightcompany'].value_counts()
    d2=pd.DataFrame(columns=['A', 'B'])
    d2['A']=list(d1.index)
    d2['B']=list(d1)
    d1=d2.sort_values(by = 'A').values.tolist()
    pie=(
    Pie(
        init_opts=opts.InitOpts(theme=ThemeType.PURPLE_PASSION#CHALK
        )
    )
    .add("",d1,center=["38%", "58%"],radius=[120, 200],)#加入数据
        .set_global_opts(title_opts=opts.TitleOpts(title="各航空公司航班数量占比"),legend_opts=opts.LegendOpts( pos_left="right", orient="vertical"))#全局设置项
        .set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {c}")))#样式设置项
    return pie

#Pie1().render('C:\\Users\\54127\\各航空公司航班数量占比.html')#保存图片为HTML网页
```

各航线词云

```python
import pandas as pd
import numpy as np
from pyecharts import options as opts#全局设置所导入的包
from pyecharts.charts import Bar, Grid, Line,WordCloud
from pyecharts.options.global_options import ThemeType
from pyecharts.globals import SymbolType

def setWordCloud():
    #统计各航线的航班数量
    data = pd.read_csv(r"C:\Users\54127\Desktop\同程旅游k.csv",encoding='utf-8')
    d1=data.loc[:,'hangxian'].value_counts()
    d2=pd.DataFrame(columns=['A', 'B'])
    d2['A']=list(d1.index)
    d2['B']=list(d1)
    d1=d2.sort_values(by = 'A').values.tolist()
    c = (
        WordCloud(init_opts=opts.InitOpts(theme=ThemeType.PURPLE_PASSION))
        .add("", d1, word_size_range=[20, 100], shape=SymbolType.DIAMOND)
        .set_global_opts(title_opts=opts.TitleOpts(title="各航线航班热度"))
        #.render("as.html")
    )
    return c
```

航班价格起伏——折线图

```python
from pyecharts.charts import Line#折线图所导入的包
from pyecharts.charts import Bar#柱状图所导入的包
from pyecharts.charts import Pie#饼图所导入的包
from pyecharts.charts import Timeline
from pyecharts import options as opts#全局设置所导入的包
from pyecharts.options.global_options import ThemeType

def setline():
    data = pd.read_csv(r"C:\Users\54127\Desktop\同程旅游票价.csv",encoding='utf-8')
    a=[0 for x in range(len(data['date']))]
    for i in range(len(data['date'])):
        a[i]=data['date'][i][5:]
    data['date']=a
    t = Timeline(init_opts=opts.InitOpts(width='1200px',height='540px',theme=ThemeType.PURPLE_PASSION))
    t.add_schema(
                 play_interval = 500,# 播放时间间隔；
                 pos_left="80px", 
                 pos_right="100px",
                 is_auto_play=True,
                 )
    for hangxian in data.hangxian.values:
        df = data[data['hangxian']==hangxian]
        line = (
            Line(
                init_opts=opts.InitOpts(
                                    theme=ThemeType.PURPLE_PASSION#CHALK
                                    )
                )#实例化Line
            .add_xaxis(list(df.date))#加入X轴数据
            .add_yaxis("最低价",list(df.min_price),markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="min")]),
             )#加入Y轴数据
            .add_yaxis("最高价",list(df.max_price),markpoint_opts=opts.MarkPointOpts(data=[opts.MarkPointItem(type_="max")]),
             )#加入Y轴数据
            .set_global_opts(title_opts=opts.TitleOpts(title="航班价格起伏", subtitle=hangxian),#设置图表标题
                            xaxis_opts=opts.AxisOpts(axislabel_opts=opts.LabelOpts(rotate=-20)),#设置x轴标签旋转角度
                            #toolbox_opts=opts.ToolboxOpts(is_show=True),
                            tooltip_opts=opts.TooltipOpts(trigger="axis"),
                            )
            .set_series_opts(label_opts=opts.LabelOpts(is_show=False))#隐藏数值
        )
        t.add(line,"{}".format(hangxian))
    #t.render('C:\\Users\\54127\\价格起伏.html')#保存图片为HTML网页
    return t

```
